```python
bank = bank.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1)
```

```python
bank['Geography'].unique()
```

```python
bank= pd.get_dummies(bank,drop_first=True)
```

```python
bank.head()
```

```python
bank['Exited'].value_counts()
```

```python
x=bank.drop('Exited',axis=1)
```

```python
y=bank['Exited']
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=0.20,random_state=42 )
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred1=log.predict=log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf =RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred2=rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

# Gradient Boosting 

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb= GradientBoostingClassifier()
```

```python
gb.fit(x_train, y_train)
```

```python
y_pred3=gb.predict(x_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
final_data=pd.DataFrame({'Models':['LogisticRegression','RandomForestClassifier','GradientBoostingClassifier'],
                            'Accuracy':[accuracy_score(y_test,y_pred1),
                                       accuracy_score(y_test,y_pred2),
                                        accuracy_score(y_test,y_pred3)]})
```
