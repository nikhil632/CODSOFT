```python
df= df[['category','gender','amt','city','state','lat','long','job','is_fraud']]
```

```python
from sklearn.utils import resample #balancing both values
df_majority=df[df.is_fraud==0]
df_minority=df[df.is_fraud==1]
print(df_majority.shape,df_minority.shape)
```

```python
df_minority_unsampled=resample(df_minority,n_samples=100000,random_state=123)
mydata3=pd.concat([df_majority[:100000],df_minority_unsampled])
print(df_minority_unsampled.shape,mydata3.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
cut_lbl=le1.fit(mydata3["category"])
mydata3["category_label"]= le1.transform(mydata3["category"])

le2=LabelEncoder()
cut_lbl=le2.fit(mydata3["gender"])
mydata3["gender_label"]= le2.transform(mydata3["gender"])

le3=LabelEncoder()
cut_lebl=le3.fit(mydata3["city"])
mydata3["city_label"]=le3.transform(mydata3["city"])

le4=LabelEncoder()
cut_lebl=le4.fit(mydata3["state"])
mydata3["state_label"]=le4.transform(mydata3["state"])

le5=LabelEncoder()
cut_lebl=le5.fit(mydata3["job"])
mydata3["job_label"]=le5.transform(mydata3["job"])
```

```python
x=dff.drop("is_fraud",axis=1)#independent variable
y=dff["is_fraud"] #Dependant variable
```

```python
from sklearn import preprocessing
#x=preprocessing.normalize(x)
#y=preprocessing scale(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
rfc= RandomForestClassifier(n_estimators=100)
lr = LogisticRegression(max_iter=1000)
dtc=DecisionTreeClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy of model:",metrics.accuracy_score(y_test,y_pred))
```
